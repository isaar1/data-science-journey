```python
import pandas as pd

df = pd.read_csv("cleaned_sales_data.csv")
```

```python
df.head()
```

## 1Ô∏è‚É£ Get all sales where Total is greater than 500

```python
high_value_sales = df[df["Total"] > 500]
high_value_sales
```

## 2Ô∏è‚É£ Get all transactions made by female customers

```python
female_customers = df[df["Gender"] == "Female"]
female_customers
```

## 3Ô∏è‚É£ Get all purchases of "Health and Beauty" products

```python
health_beauty_sales = df[df["Product line"] == "Health and beauty"]
health_beauty_sales
```

## 1Ô∏è‚É£ Find the total sales per product line

```python
total_sales_per_product = df.groupby("Product line")["Total"].sum()

total_sales_per_product
```

## 2Ô∏è‚É£ Find the average rating per city

```python
avg_rating_per_city = df.groupby("City")["Rating"].mean()

avg_rating_per_city
```

## 3Ô∏è‚É£ Find the total quantity sold per branch

```python
total_quantity_per_branch = df.groupby("Branch")["Quantity"].sum()

total_quantity_per_branch
```
